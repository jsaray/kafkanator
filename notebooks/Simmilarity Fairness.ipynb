```python
import pandas as pd

# Specify the path to the CSV file
file_path = '/home/jsaray/Data Science/Fairness/adult/adult2.csv'

# Read the CSV file into a DataFrame
data = pd.read_csv(file_path,sep=',',
        names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','gender','capital-loss','capital-gain','hours-per-week','native-country','income'],
        dtype={'age':'int64','workclass':str,'fnlwgt':'int64','education':str,'education-num':'int64','marital-status':str,'occupation':str,'relationship':str,'race':str,'gender':str, 'capital-loss':'int64','capital-gain':'int64',
               'hours-per-week':'int64','native-country':'str','income':str})
```

```python
def categorical_simmilarity_distance(ind1,ind2):
    distanceArray = []
    customized_attr_types = { 'age':'num', 
               'workclass':'cat',
               'education': 'cat' ,
               'education-num' :'num'  ,
               'marital-status':'cat'  ,
               'occupation':'cat' ,
               'relationship':'cat' ,
                'race':'cat' ,
                'capital-loss':'num' ,
                'capital-gain':'num',
                'hours-per-week':'num' ,
                'native-country':'cat',
                'income':'target'}
    for (k,v) in customized_attr_types.items():
        if v == 'cat':
            toApp = 0 if ind1[k] == ind2[k] else 1
            distanceArray.append( toApp )
        elif v == 'num':
            toApp = abs( ind1[k] - ind2[k] )
            distanceArray.append( toApp )
        elif v == 'target':
            samePrediction = True if ind1[k] == ind2[k] else False
    if samePrediction == None:
        assert('You must put a target')
    if len(distanceArray) == 0:
        assert('You must add your table columns')
    return distanceArray,samePrediction
```

```python
def simmilarity_fairness_hash( df_prot_attr1 , df_prot_attr2, simmilarity_distance):
    hDict = {}
    for i,w in df_prot_attr1.iloc[0:100].iterrows():
        #print ( ' checking more simmilar individual for women in index ' , w)
        for j,h in df_prot_attr2.iloc[0:100].iterrows():
            reto = simmilarity_distance(w,h)
            #print ( ' i ', i,' j ', j , ' reto ', reto ,  ' fnlwgt W ' , w['fnlwgt'] , ' fnlwgt H ' , h['fnlwgt'] )
            d = sum(reto[0])
            hDict[(i,j)] = (d,reto[1])

    sor = sorted(hDict.items(), key=lambda row: row[1][0], reverse=False)
    return sor
```

```python
wo = data[data['gender'] == ' Male' ].reset_index()
wo = wo.drop(['index'],axis=1)
ma = data[data['gender'] == ' Female'  ].reset_index()
ma = ma.drop(['index'],axis=1)

hsh = simmilarity_fairness_hash( wo,ma,categorical_simmilarity_distance )
```
